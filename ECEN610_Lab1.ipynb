```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lfilter, tf2zpk

# Simulate an input signal (a step function)
n = np.linspace(0, 50, 51)
x_iir = np.ones_like(n)

# Define an unstable IIR filter by choosing a pole outside the unit circle
b_unstable = [1]
a_unstable = [0.7, 1.5, 1]  # This filter has a pole outside the unit circle

# Apply the unstable IIR filter to the input signal
y_unstable = lfilter(b_unstable, a_unstable, x_iir)

# Plot the input signal (step function)
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(n, x_iir, label='Input Signal (Step Function)', color='blue')
plt.title('Input Signal')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

# Plot the output of the unstable IIR filter
plt.subplot(1, 2, 2)
plt.plot(n, y_unstable, label='Unstable IIR Filter Output', color='red')
plt.title('Unstable IIR Filter Response')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# Get poles and zeros of the IIR filter
zeros, poles, _ = tf2zpk(b_unstable, a_unstable)

# Plot the poles and zeros on the z-plane
plt.figure(figsize=(6, 6))

# Plot the zeros and poles
plt.scatter(np.real(zeros), np.imag(zeros), marker='o', label='Zeros', color='blue')  # Zeros
plt.scatter(np.real(poles), np.imag(poles), marker='x', label='Poles', color='red')  # Poles

# Mark the unit circle (to check stability)
unit_circle = plt.Circle((0, 0), 1, color='black', fill=False, linestyle='--', linewidth=1)
plt.gca().add_artist(unit_circle)

# Set plot labels and title
plt.title('Poles and Zeros of the Unstable IIR Filter with Unit Circle')
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True)
plt.legend()

# Adjust the axis limits to ensure the unit circle is visible
plt.xlim(-2, 2)
plt.ylim(-2, 2)

# Show the z-plane plot
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
f_cos = 300000000  # Frequency of the cosine wave (in Hz)
f_samp = 800000000  # Sampling rate (in Hz), should be >= 2 * f_cos for Nyquist criterion
t_max = 10/300000000  # Duration of the signal (in seconds)

# Generate the continuous time (high-resolution) time vector
t_continuous = np.linspace(0, t_max, 1000)  # High-resolution time vector
cos_wave = np.cos(2 * np.pi * f_cos * t_continuous)  # Continuous cosine wave

# Generate the sampled time vector
t_sampled = np.arange(0, t_max - 1/f_samp, 1/f_samp)  # Sampled time vector
cos_wave_sampled = np.cos(2 * np.pi * f_cos * t_sampled)  # Sampled cosine wave

# Plot both the continuous and sampled signals
plt.figure(figsize=(10, 6))

# Plot the continuous signal
plt.plot(t_continuous, cos_wave, label='Continuous Cosine Wave', color='blue', lw=2)

# Plot the sampled signal
plt.stem(t_sampled, cos_wave_sampled, label='Sampled Cosine Wave', linefmt='red', markerfmt='ro', basefmt=" ")

plt.title('Sampling of a Cosine Wave')
plt.xlabel('Time [seconds]')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# Show the plot before printing MSE
plt.show()

# Define the sinc function
def sinc(x):
    return np.sinc(x)

# Define the interpolation function
def interpolate(x_samples, Ts, t_values):
    # Initialize the interpolated signal
    interpolated_signal = np.zeros_like(t_values)

    # Loop over the samples and accumulate their contributions
    for n, x_n in enumerate(x_samples):
        # Shift the sinc function by n * Ts and evaluate at the specified times
        interpolated_signal += x_n * sinc((t_values - n * Ts) / Ts)

    return interpolated_signal

interpolated_x1 = interpolate(cos_wave_sampled, 1/f_samp, t_continuous)

# Plot the interpolated signal
plt.plot(t_continuous, interpolated_x1, label='Interpolated Cosine Wave', color='blue', lw=2)
plt.title('Interpolated Cosine Wave')
plt.xlabel('Time [seconds]')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

# Define a new set of sampled points with shifted time
t_sampled_new = np.arange(1/(f_samp*2), t_max - 1/(f_samp*2), 1/f_samp)  # Sampled time vector
cos_wave_sampled_new = np.cos(2 * np.pi * f_cos * t_sampled_new)  # Sampled cosine wave

# Plot the continuous and new sampled signals
plt.figure(figsize=(10, 6))

plt.plot(t_continuous, cos_wave, label='Continuous Cosine Wave', color='blue', lw=2)
plt.stem(t_sampled_new, cos_wave_sampled_new, label='New Sampled Cosine Wave', linefmt='red', markerfmt='ro', basefmt=" ")

plt.title('Sampling of a Cosine Wave with Shifted Time')
plt.xlabel('Time [seconds]')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# Show the plot before printing MSE
plt.show()

# Reconstruct using sinc interpolation with the new samples
interpolated_x1_new = interpolate(cos_wave_sampled_new, 1/f_samp, t_continuous)

# Plot the interpolated shifted signal
plt.plot(t_continuous, interpolated_x1_new, label='Interpolated Shifted Cosine Wave', color='blue', lw=2)
plt.title('Interpolated Shifted Cosine Wave')
plt.xlabel('Time [seconds]')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

# Define the Mean Squared Error function
def mean_squared_error(array1, array2):
    if len(array1) != len(array2):
        raise ValueError("Arrays must have the same length")
    mse = np.mean((array1 - array2) ** 2)
    return mse

# Compute the MSE between the interpolated signals
mse = mean_squared_error(interpolated_x1_new, interpolated_x1)
print(f"Mean Squared Error: {mse}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import lfilter, tf2zpk

# Simulate an input signal (a step function)
n = np.linspace(0, 50, 51)
x_fir = np.ones_like(n)

# Define the FIR filter (1 + z^-1 + z^-2 + z^-3 + z^-4)
b_fir = [1, 1, 1, 1, 1]  # FIR filter coefficients
a_fir = [1]  # FIR filters have no feedback, so denominator is just [1]

# Apply the FIR filter to the input signal
y_fir = lfilter(b_fir, a_fir, x_fir)

# Plot the input signal (step function)
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(n, x_fir, label='Input Signal (Step Function)', color='blue')
plt.title('Input Signal')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

# Plot the output of the FIR filter
plt.subplot(1, 2, 2)
plt.plot(n, y_fir, label='FIR Filter Output', color='green')
plt.title('FIR Filter Response')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# Get poles and zeros of the FIR filter
zeros, poles, _ = tf2zpk(b_fir, a_fir)

# Plot the poles and zeros on the z-plane
plt.figure(figsize=(6, 6))

# Plot the zeros and poles
plt.scatter(np.real(zeros), np.imag(zeros), marker='o', label='Zeros', color='blue')  # Zeros
plt.scatter(np.real(poles), np.imag(poles), marker='x', label='Poles', color='red')  # Poles

# Mark the unit circle (to check stability)
unit_circle = plt.Circle((0, 0), 1, color='black', fill=False, linestyle='--', linewidth=1)
plt.gca().add_artist(unit_circle)

# Set plot labels and title
plt.title('Poles and Zeros of the FIR Filter with Unit Circle')
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True)
plt.legend()

# Adjust the axis limits to ensure the unit circle is visible
plt.xlim(-2, 2)
plt.ylim(-2, 2)

# Show the z-plane plot
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
